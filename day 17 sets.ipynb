```python
#sets
#set is a unordered collection of different data type unique elements ,it is mutable and represents with '{}
```

```python
#set allows unique values
#set unordered
s1={1,2,3,4,'hi',1,2,3,'a'}
print(s1)

```

```python
#set not allows list as elements
s1={1,2,3,4,'hi',1,2,3,'a',[1,2,3,4]}
print(s1)
```

```python
#set not allows distionary as elements
s1={1,2,3,4,'hi',1,2,3,'a',{1:2}}
print(s1)
```

```python
## set functions
#adding elements to set
#add(ele):it add the element to set if element is not in set otherwice nothing
#update(iterator): it add the all elements in iterator
```

```python
s1={1,2,3,4}
s1.add(10)
s1.add(1)
s1.update((5,3,5))
print(s1)
```

```python
# remove operation
s1={1,2,3,4}
s1.remove(4)
print(s1)
```

```python
# pop operation in a set
s1={1,2,3,4}
s1.pop()
print(s1)
```

```python
# discard operation in aset
s1={1,2,3,4}
s1.discard(3)
print(s1)
```

```python
# union
s1={1,2,3,4,5,6}
s2={4,5,6}
s3=s1.union(s2)
print(s3)
print(s1.intersection(s2))
print(s1.difference(s2))
print(s1-s2)
print(s1.symmetric_difference(s2))
print({1}.issubset(s1))
```
