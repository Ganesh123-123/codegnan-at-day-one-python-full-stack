```python
#updating the values and items
details={'name':'ganesh','batch':47,'cource':'python'}
details['name'] ='babu'
print(details)
# assining new item into dictionary
#dict_obj[key] =value
details['age']=50
print(details)
```

```python
# dictionary funmctions
#accessing values
#get{key,default_value): it returns value if key is present in dictionary atherwise default value
details={'name':'babu','batch':47,'cource':'python','age':50}
print(details.get('name'))
print(details.get('percentage'))
print(details.get('percentage',90))


```

```python
#updating or adding items to distionary 
#update(new_dictionary):it updates with new_dictionary items
details={'name':'ganesh','batch':47,'cource':'python','age':50,'percentage':85}
marks={'python':90,'soft skills':85,'aptitude':78,'percentage':90}
details.update(marks)
print(details)
print(marks)
```

```python
# get dictionary keys and values 
#keys()- it returns list of all dictionary keys
#values()-it returns list of all dictionary values
#items()-it returns all dictionary items as list of key-value pairs
details={'name':'ganesh','batch':47,'cource':'python','age':50,'percentage':85,}
print(details.keys())
print(details.values())
print(details.items())
print(type(details.keys()))
print(type(details.values()))
print(type(details.items()))
print(list(details.keys()))




```

```python
# deleting items from distionary
#pop(key)- it return value and removes the specified key item from dictionary
#popitem()- it return and removes the last item from dictionary
details={'name':'ganesh','batch':47,'cource':'python','age':70}
poped_value =details.pop('age')
print(poped_value)
print(details)
poped_value = details.pop('age',None)
print(poped_value)
print(details)
item=details.popitem()
print(item)
print(details)
```
