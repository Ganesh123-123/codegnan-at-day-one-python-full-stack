```python
#USING FUNCTIONS
def add(a,b):
    r=a+b
    return r
def sub(a,b):
    s=a-b
    return s
def mul(a,b):
    t=a*b
    return t
def div(a,b):
    if b==0 :
        return "zero division not possible"
    u=a/b
    return u

if __name__ == "__main__" :
    print("welcome to small level calculator") 
n1=int(input())
n2=int(input())
print("1 for addition /n 2 for subtraction /n 3 for multiplication /n 4 for division  ")
choice=int(input("enter your choice"))
if choice ==1:
    print("sum of 2 numbers is",add(n1,n2))
elif choice ==2:
     print("subtraction of 2 numbers is",sub(n1,n2))
elif choice ==3:
     print("multiplication  of 2 numbers is",mul(n1,n2))
elif choice ==4:
     print("division of 2 numbers is",div(n1,n2))
```
