```python
#convert list of numbers into even list odd list
l=list(map(int,input().split()))
even_nums=[]
odd_nums=[]
for num in l:
    if num % 2:
        odd_nums.append(num)
    else:
        even_nums.append(num)


print(even_nums)
print(odd_nums)
       
```

```python
#Marks analyze
#in listcontans list of students marks, in that find the pass percentage and fail percentage
#pass maeks is 35+
#fail is below 35

marks = list(map(int, input().split()))
pass_count = 0
fail_count = 0

for num in marks:
    if num >= 35:
        pass_count += 1
    else:
        fail_count += 1
pass_per = (pass_count/len(marks))*100
fail_per = (fail_count/len(marks))*100
print("pass % = ",pass_per)
print("fail % = ", fail_per)
```
